```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
f = open("D:\ML project\Whatapp_chat_analyser\chats/WhatsApp Chat with Didi.txt","r",encoding = "utf-8")
```

```python
data = f.read()
```

```python
pattern =r"\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s?(?:AM|PM)\s-\s"
```

```python
message = re.split(pattern,data)[1:] # take second part
```

```python
messages = [m.strip() for m in message if m.strip()] # Remove empty spaces
```

```python
messages[1:20]
```

```python
data = data.replace("\u202f", " ")
dates = re.findall(pattern,data)
```

```python
dates[1:10]
```

```python
df = pd.DataFrame({'datetime_str': dates, 'message': messages})
```

```python
df.head()
```

```python
df['datetime_str'] = df['datetime_str'].str.replace(" - ", "")  #regex=False
df['datetime'] = pd.to_datetime(df['datetime_str'], format="%m/%d/%y, %I:%M %p") ## %I->hour %M ->minute %P-> AM/PM
```

```python
df.drop(columns='datetime_str', inplace=True)
```

```python
users =[]
messages = []
for message in df["message"]:
    entry = re.split(r"([\w\W]+?):\s",message)
        
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append("goup_notification")
        messages.append(entry[0])

df["user"] = users
df["messages"] = messages
df.drop(columns=["message"],inplace=True)
```

```python
df.head()
```

```python
df["year"] = df["datetime"].dt.year
df["month_num"] = df["datetime"].dt.month
df["month"] = df["datetime"].dt.month_name()
df["day"] = df["datetime"].dt.day
df["date"] = df["datetime"].dt.date
df["day_name"] = df["datetime"].dt.day_name()
df["hour"] = df["datetime"].dt.hour
df["minute"] = df["datetime"].dt.minute
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
words = []
for message in df['messages']:
    for word in message.lower().split(): # Convert to lowercase
        if word not in stop_words and word.isalpha(): # remove stop_words and isalpha() remove punctuation (!@#$%&()/+-)
            words.append(word)
```

```python
words
```

```python
from collections import Counter
most_commonwords_df = pd.DataFrame(Counter(words).most_common(20))
```

```python
most_commonwords_df
```

```python
sns.barplot(x=most_commonwords_df[1],y = most_commonwords_df[0],orient='h',palette = "plasma")
```

```python
period_time = []
for hour in df[["day_name","hour"]]["hour"]:
    if hour == 23:
        period_time.append(str(hour)+"-"+ str("00"))
    elif hour == 0:
        period_time.append(str("00")+"-"+ str(hour+1))
    else:
        period_time.append(str(hour)+"-"+ str(hour+1))
    
df["period_time"] = period_time
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
import emoji as em
```

```python
df["demojize_messages"] = df["messages"].apply(lambda x: em.demojize(str(x), language="en"))

# Initialize Sentiment Analyzer
sia = SentimentIntensityAnalyzer()

# Get compound sentiment score
df["sentiment_score"] = df["demojize_messages"].apply(lambda x: sia.polarity_scores(str(x))["compound"])

# Classify sentiment as Negative, Neutral, Positive
df["sentiment"] = pd.cut(
    df["sentiment_score"],
    bins=[-1, -0.05, 0.05, 1],
    labels=["Negative", "Neutral", "Positive"]
)
```
